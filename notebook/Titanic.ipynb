# Imports

```python
import numpy as np
import pandas as pd 
```

# Load data

```python
train_data = pd.read_csv("../data/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("../data/test.csv")
test_data.head()
```

# Explore data

```python
print(train_data.shape)
test_data.shape
```

```python
train_data.describe(include='all').T
```

## Missing Values

```python
print(train_data.isnull().sum().sort_index()/len(train_data))
```

```python
print(test_data.isnull().sum().sort_index()/len(test_data))
```

```python
train_data['Age'] = train_data['Age'].interpolate()
test_data['Age'] = test_data['Age'].interpolate()
test_data['Fare'] = test_data['Fare'].interpolate()
```

# Model

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
```

```python
X.dtypes
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('my_submission.csv', index=False)
print("Your submission was successfully saved!")
```
